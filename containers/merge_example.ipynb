```python
import pandas as pd
import os
```

```python
cells = [item for item in os.listdir('tmpresult') if os.path.isdir(os.path.join('tmpresult', item))]
cells
```

```python
dataframes = []
for cell in cells:
    df = pd.read_csv(os.path.join('tmpresult', cell, 'carnival.csv'))
    df.set_index(df.columns[0], inplace = True)
    df.rename(columns={df.columns[0]: cell }, inplace = True)
    dataframes.append(df)
df_result = pd.concat(dataframes, axis=1, verify_integrity=True)
df_result
```

```python
t = df_result.abs().sum(axis=1)
df_result.loc[t>0, :]
```

```python
df_result.loc[t>1, :]
```

```python
with open('genelist.txt', mode='wt', encoding='utf-8') as f:
    f.write('\n'.join(df_result.loc[t>1, :].index.to_list()))
```
